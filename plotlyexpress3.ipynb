```python
import plotly
import plotly.express as px
```

```python
def fetch_financial_data(company = "AMZN"):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = "stooq")

df = fetch_financial_data()
df = df.reset_index()
df.head()
```

```python
px.area(data_frame = df, x = "Date", y = "Close", log_y = True)
```

```python
from seaborn import load_dataset
df = load_dataset("flights")
df.head()
```

```python
px.bar(data_frame = df, x = "year", y = "passengers", color = "year")
```

```python
px.bar(data_frame = df, y = "year", x = "passengers", color = "year", orientation = "h")
```

```python
px.histogram(df, x = "passengers", nbins = 50)
```

```python
import pandas as pd
df = pd.read_csv("https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/plotly-course/us-cities-top-1k.csv")
df.head()
```

```python
fig = px.scatter_mapbox(df, lat = "lat", lon = "lon", hover_name = "City", hover_data = ["State", "Population"], zoom = 3)
fig.update_layout(mapbox_style = "carto-positron", margin = {"r" : 10, "t" : 10, "l" : 10, "b" : 10})
fig.show()
```
